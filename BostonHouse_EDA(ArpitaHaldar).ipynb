# Data Manipulation:

**Add a new column to the DataFrame that categorizes the ‘medv’ (median value of owner-occupied homes) column into ‘Low’, ‘Medium’, and ‘High’.**

```python
tata['MEDV'].value_counts()
```

```python
tata['MEDV'] = pd.cut(data['medv'], bins=[0,15,30,50], labels=['Low', 'Medium', 'High'])

tata.head()
```

**Replace the ‘chas’ column (Charles River dummy variable) with ‘Yes’ if chas = 1
and ‘No’ if chas = 0.**

```python
tata['CHAS'] = tata['CHAS'].replace({1: 'Yes', 0: 'No'})
tata.head()
```

### **Heat Map:**

```python
h_m = pd.crosstab(tata['MEDV'], tata['CHAS'])

h_m
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(h_m, annot=True, fmt='d', cmap='Spectral')
plt.title('Relationship between Median Value(MEDV) and Charles River(CHAS)', fontsize=14)
plt.xlabel('CHAS', fontsize=12)
plt.ylabel('MEDV', fontsize=12)
plt.show()
```

**Rename the column ‘RM’ to ‘Room’.**

```python
tata.rename(columns={'RM': 'Room'}, inplace=True)
tata.head()
```

# Data Analysis:

**Find out the average number of rooms (‘rooms’ column) per dwelling for each
category of ‘medv’ (Low, Medium, High).**

```python
room_medv = tata.groupby('MEDV')['Room'].mean()

room_medv
```

### **Bar Plot:**

```python
plt.figure(figsize=(8,5))
sns.barplot(x=room_medv.index, y=room_medv.values, palette='rocket')

plt.title('Average Number of Rooms per Dwelling by Median Value', fontsize=16)
plt.xlabel('Median Value', fontsize=14)
plt.ylabel('Average Number of Rooms', fontsize=14)

plt.show()
```

**Find out the percentage of houses that bound the Charles River.**

```python
total_houses = len(tata)
h_b_charles = len(tata[tata['CHAS'] == 'Yes'])
percentage = (h_b_charles / total_houses) * 100

print(f"Percentage of houses that bound the Charles River : {percentage:.2f}%")
```

# NumPy Operations:

**Convert the ‘age’ column (proportion of owner-occupied units built prior to 1940) from the DataFrame to a NumPy array and perform the following operations:**

```python
age_ary = np.array(tata['AGE'])
age_mn = np.mean(age_ary)
age_std = np.std(age_ary)

print("Mean: ", age_mn)
print("Standard Deviation: ", age_std)
```

```python
#2. Normalize the array.

age_norz = (age_ary - age_mn)/age_std

print("Normalized array: ", age_norz)
```

# Data visualization:

### **Density Plot:**

```python
plt.figure(figsize=(6,4))
tata['TAX'].plot.density()

plt.xlabel('TAX', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.title('Density Plot of TAX', fontsize=14)

plt.grid(True)
plt.legend()
plt.show()
```

### **Pie Chart:**

```python
h_count = tata['CHAS'].value_counts()

plt.figure(figsize=(6, 4))
plt.pie(h_count, labels=h_count.index, autopct="%1.1f%%", colors=['lightcoral', 'teal'])

plt.title('Distribution of Houses by Charles River', fontsize=14)
plt.legend(loc='lower right')

plt.show()
```

### **Distribution Plot:**

```python
plt.figure(figsize=(8, 6))
sns.displot(tata['AGE'], bins=10, color='Purple', kde=True)

plt.xlabel('AGE', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Distribution of AGE', fontsize=16)

plt.show()
```

### **Box Plot:**

```python
sns.boxplot(x='MEDV', y='AGE', data=tata)

plt.xlabel('MEDV', fontsize=14)
plt.ylabel('AGE', fontsize=14)
plt.title('Box Plot of AGE by MEDV', fontsize=16)

plt.show()
```

### **Scatter Plot:**

```python
plt.figure(figsize=(6, 5))
sns.scatterplot(x='AGE', y='TAX', data=tata, hue='MEDV',palette='viridis')

plt.xlabel('AGE', fontsize=12)
plt.ylabel('TAX', fontsize=12)
plt.title('Scatter Plot of AGE vs TAX', fontsize=14)

plt.show()
```
